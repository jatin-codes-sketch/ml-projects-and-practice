```python
# Sigmoid funtion is defined as:
# f(x) = 1 / (1 + exp(-x))
# # This function is used in logistic regression and neural networks.
# # The sigmoid function maps any real-valued number into the range (0, 1).
# # The derivative of the sigmoid function is:
# f'(x) = f(x) * (1 - f(x))
# # This script implements the sigmoid function and its derivative.
```

```python
from sklearn.datasets import make_classification 
import numpy as np 
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=32,hypercube=False,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(10000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
        
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,7))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(10000):
        j=np.random.randint(0,100)
        y_hat=sigmoid(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]
        
```

```python
intercept_,coef_=perceptron(X,y)
```
